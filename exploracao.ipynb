```python
import os
import re
import pandas as pd
from datetime import datetime

# Caminho base do projeto (ajustado conforme estrutura)
caminho_base = os.path.join(os.path.dirname(os.getcwd()), 'data_raw')
pastas_servidores = {
    's5': os.path.join(caminho_base, 'servidor5'),
    's6': os.path.join(caminho_base, 'servidor6')
}

dados = []

def extrair_info(linha, servidor, ano):
    try:
        if not linha.startswith("-rw"):
            return None

        partes = linha.split(maxsplit=8)
        if len(partes) < 9:
            return None

        tamanho = int(partes[4])
        nome_arquivo = partes[8]

        base_match = re.search(r'arq_(x2[^_]+)', nome_arquivo)
        data_match = re.search(r'_(\d{4})', nome_arquivo)  # extrai apenas o dia (0210)

        if not (base_match and data_match):
            return None

        base = base_match.group(1)
        data_raw = data_match.group(1)

        # Formata 0210 → 10-02-2024 (usando ano fixo por enquanto)
        data = datetime.strptime(data_raw + str(ano), '%m%d%Y').strftime('%d-%m-%Y')

        return {
            'Base': base,
            'Servidor': servidor,
            'Data': data,
            'Tamanho (bytes)': tamanho
        }
    except Exception as e:
        print(f"Erro ao processar linha: {linha}\n{e}")
        return None

# Processa os arquivos dos servidores
for servidor, pasta in pastas_servidores.items():
    if not os.path.exists(pasta):
        print(f"❌ Diretório não encontrado: {pasta}")
        continue

    for arquivo in os.listdir(pasta):
        if arquivo.endswith('.txt'):
            caminho_arquivo = os.path.join(pasta, arquivo)

            # Extrai o ano a partir do nome do arquivo
            ano_match = re.search(r'lista(\d{4})\d{4}\.txt', arquivo)
            if not ano_match:
                print(f"⚠️ Ano não encontrado no nome do arquivo: {arquivo}")
                continue
            ano = int(ano_match.group(1))  # ex: 2024

            try:
                with open(caminho_arquivo, 'r', encoding='utf-8') as f:
                    for linha in f:
                        info = extrair_info(linha.strip(), servidor, ano)
                        if info:
                            dados.append(info)
            except Exception as e:
                print(f"⚠️ Erro ao abrir {caminho_arquivo}: {e}")


# Cria o DataFrame
df = pd.DataFrame(dados)

if df.empty:
    print("🚫 Nenhum dado extraído. Verifique o padrão dos nomes nos arquivos.")
else:
    print("✅ Dados extraídos com sucesso!")
    print(df.head())

    # Agrupa e exporta os dados
    df_agrupado = df.groupby(['Base', 'Servidor', 'Data']).agg({'Tamanho (bytes)': 'sum'}).reset_index()

    caminho_saida = os.path.join(os.getcwd(), 'resultado_agrupado.csv')
    df_agrupado.to_csv(caminho_saida, index=False, encoding='utf-8')
    print(f"📁 CSV salvo em: {caminho_saida}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Carregamento do CSV
df = pd.read_csv('resultado_agrupado.csv', sep=',', encoding='utf-8')

# Conversão da coluna de data
df['Data'] = pd.to_datetime(df['Data'], format='%d-%m-%Y')

# Conversão de Bytes para GB (1 GB = 1024^3 Bytes)
df['Tamanho (GB)'] = df['Tamanho (bytes)'] / (1024**3)

# Filtragem opcional se houver dados reais e projetados
if 'Tipo' in df.columns:
    df_real = df[df['Tipo'] == 'Real'].copy()
else:
    df_real = df.copy()

# Análise por servidor (S5 e S6)
for servidor in ['s5', 's6']:
    df_servidor = df_real[df_real['Servidor'] == servidor].copy()
    
    # Separar por base (assumindo que há múltiplas bases dentro de cada servidor)
    for base in df_servidor['Base'].unique():
        base_df = df_servidor[df_servidor['Base'] == base].copy()
        
        # Ordena cronologicamente
        base_df = base_df.sort_values('Data')
        base_df['dias'] = (base_df['Data'] - base_df['Data'].min()).dt.days
        
        # Regressão linear para projeção
        X = base_df[['dias']]
        y = base_df['Tamanho (GB)']
        modelo = LinearRegression().fit(X, y)

        # Projeção para os próximos 6 meses a cada 7 dias
        dias_futuros = np.arange(base_df['dias'].max() + 1, base_df['dias'].max() + 180, 7)
        previsoes = modelo.predict(dias_futuros.reshape(-1, 1))
        datas_futuras = base_df['Data'].min() + pd.to_timedelta(dias_futuros, unit='D')

        projecao_df = pd.DataFrame({
            'Base': base,
            'Servidor': servidor,
            'Data': datas_futuras,
            'Tamanho (GB)': previsoes,
            'Tipo': 'Projeção'
        })

        # Junta com dados reais
        base_df['Tipo'] = 'Real'
        resultado_df = pd.concat([base_df[['Base', 'Servidor', 'Data', 'Tamanho (GB)', 'Tipo']], projecao_df])

        # Exporta o resultado
        output_dir = 'notebooks/projecoes'
        os.makedirs(output_dir, exist_ok=True)
        nome_arquivo = os.path.join(output_dir, f'crescimento_{servidor}_{base}.csv')
        resultado_df = resultado_df.sort_values('Data')
        resultado_df.to_csv(nome_arquivo, index=False, encoding='utf-8')
        print(f'📁 Arquivo gerado: {nome_arquivo}')
        print("Local absoluto:", os.path.abspath(nome_arquivo))
```

```python
import os
import re
import pandas as pd

# Caminhos
base_path = '../data_raw/servidor5'
saida_excel = '../data_raw/saida.xlsx'
log_path = '../data_raw/log_processamento.txt'

# Regex para capturar linhas válidas (com ou sem 'arq' após lwdump)
line_regex = re.compile(
    r'^-rw-r--r--\.\s+\d+\s+\w+\s+\w+\s+(\d+)\s+\w+\s+\d+\s+\d+:\d+\s+(s\d)_lwdump(_arq)?_([a-z0-9]+)_\d+[-\d+]*\.7z\.\d+$',
    re.IGNORECASE
)

# Bases permitidas
bases_validas = ('x2', 'x1', 'midas2', 'turiion')

dados = []
arquivos_processados = []

# Processar cada arquivo .txt
for filename in os.listdir(base_path):
    if filename.endswith('.txt'):
        data_do_arquivo = filename.replace('lista', '').replace('.txt', '')
        data_formatada = f"{data_do_arquivo[:4]}-{data_do_arquivo[4:6]}-{data_do_arquivo[6:]}"
        
        try:
            with open(os.path.join(base_path, filename), 'r') as file:
                for line in file:
                    match = line_regex.match(line.strip())
                    if match:
                        tamanho, servidor, arq_flag, base = match.groups()
                        # Filtrar bases válidas
                        if not base.lower().startswith(bases_validas):
                            continue
                        tipo_arquivo = 'arq' if arq_flag else 'normal'
                        dados.append({
                            'Data': data_formatada,
                            'Servidor': servidor,
                            'Base': base,
                            'Tamanho (MB)': int(tamanho) / (1024*1024),
                            'Tipo': tipo_arquivo
                        })
            arquivos_processados.append(f'✅ {filename}')
        except Exception as e:
            arquivos_processados.append(f'❌ {filename} - Erro: {e}')

# Salvar log de arquivos
with open(log_path, 'w') as log_file:
    log_file.write('\n'.join(arquivos_processados))

# Criar DataFrame e processar os dados
df = pd.DataFrame(dados)

# Agrupamento geral (todos os arquivos)
df_agrupado = df.groupby(['Data', 'Servidor', 'Base'], as_index=False)['Tamanho (MB)'].sum()
df_ordenado = df_agrupado.sort_values(by=['Base', 'Data'])

# Totais por Base (todos os arquivos)
df_totais = df_ordenado.groupby('Base')['Tamanho (MB)'].sum().reset_index()

# Tamanhos separados por tipo
df_arq = df[df['Tipo'] == 'arq'].groupby(['Base', 'Data'])['Tamanho (MB)'].sum().reset_index()
df_normal = df[df['Tipo'] == 'normal'].groupby(['Base', 'Data'])['Tamanho (MB)'].sum().reset_index()

# Crescimento por Base e Data (apenas arquivos normais)
df_crescimento = df_normal.copy()
df_crescimento = df_crescimento.sort_values(['Base', 'Data'])
df_crescimento['Crescimento (%)'] = df_crescimento.groupby('Base')['Tamanho (MB)'].pct_change() * 100
df_crescimento['Diferença (MB)'] = df_crescimento.groupby('Base')['Tamanho (MB)'].diff()

# Exportar tudo em um único Excel com múltiplas abas
with pd.ExcelWriter(saida_excel, engine='openpyxl') as writer:
    df_ordenado.to_excel(writer, sheet_name='Resumo por Base', index=False)
    df_totais.to_excel(writer, sheet_name='Totais por Base', index=False)
    df_crescimento.to_excel(writer, sheet_name='Crescimento (%)', index=False)
    df_arq.to_excel(writer, sheet_name='Apenas arq', index=False)
    df_normal.to_excel(writer, sheet_name='Apenas normal', index=False)

print(f'✅ Relatório completo exportado para: {saida_excel}')
```
